```python
import time, random,turtle

delay = 0.1
wn = turtle.Screen()
wn.title("January 7 exercise")

#wn.bgcolor("white")
wn.bgpic("sand2.png")
wn.setup(width=600, height=600)
wn.tracer(0) 
head = turtle.Turtle()
head.speed(0)
head.shape("turtle")
head.color("green")
head.penup()
head.goto(0, 0)
head.direction = "stop"

food = turtle.Turtle()
food.speed(0)
food.shape("circle")
food.color("red")
food.penup()
food.goto(0, 100)

bird= turtle.Turtle()


score = 1
scoreb= turtle.Turtle()
scoreb.speed(0)
scoreb.penup()
scoreb.goto(0,200)
segments = []

def move():
	if head.direction == "up":
		y = head.ycor()
		head.sety(y + 20)
		head.direction= "stop"

	if head.direction == "down":
		y = head.ycor()
		head.sety(y - 20)
		head.direction= "stop"
	if head.direction == "left":
		x = head.xcor()
		head.setx(x - 20)
		head.direction= "stop"
	if head.direction == "right":
		x = head.xcor()
		head.setx(x + 20)
		head.direction="stop"
def bird_attack():
	bird.speed(0)
	bird.shapesize(1)
	bird.shape("arrow")
	bird.color("black")
	bird.penup()
	#random starting point
	x = random.randint(-290, 290)
	y = random.randint(-290, 290)
	bird.goto(x,y)
	
def go_up():
	head.direction = "up"
def go_down():
	head.direction = "down"
def go_left():
	head.direction = "left"
def go_right():
	head.direction = "right"

wn.listen()
wn.onkeypress(go_up, "Up")
wn.onkeypress(go_down, "Down")
wn.onkeypress(go_left, "Left")
wn.onkeypress(go_right, "Right")

def check_collision_borders():
	if head.xcor() > 290 or head.xcor() < -290 or head.ycor() > 290 or head.ycor() < -290:
		time.sleep(1)
		head.goto(0, 0)
		head.direction = "stop"				
		for i in segments:
			i.getturtle().reset()
		segments.clear()
		score = 0

while True:
	wn.update()
	check_collision_borders()
	scoreb.clear()
	scoreb.write("Your turtle family: "+str(score), move=False, align="left", font=("Arial", 20, "normal"))
	if head.distance(bird) < 20:
		segments.pop()
		score= score-1
	if score==2:
		bird_attack()
		scoreb.clear()
		scoreb.write("BATS!!", move=False, align="left", font=("Arial", 20, "normal"))
		time.sleep(delay)
		scoreb.clear()

	if head.distance(food) < 20:
		x = random.randint(-290, 290)
		y = random.randint(-290, 290)
		food.goto(x, y)

		new_segment = turtle.Turtle()
		new_segment.speed(0)
		new_segment.shape("turtle")
		new_segment.color('grey')
		new_segment.penup()
		segments.append(new_segment)
		score = score + 1


	for index in range(len(segments)-1, -1,-1):
		x = segments[index-1].xcor()
		y = segments[index-1].ycor()
		segments[index].goto(x, y)

	if len(segments) > 0:
		x = head.xcor()
		y = head.ycor()
		segments[0].goto(x, y)

	move()
	time.sleep(delay)

wn.mainloop()
```
